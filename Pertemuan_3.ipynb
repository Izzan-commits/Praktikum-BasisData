# üìò Rangkuman Pengoperasian SQL Menggunakan CLI (MySQL)

Dokumen ini menjelaskan langkah-langkah dasar dalam mengelola database menggunakan Command Line Interface (CLI) pada MySQL.

<br>

---

## üìÅ 1. Membuat Database

Pembuatan database dilakukan untuk menyediakan wadah atau tempat penyimpanan berbagai tabel yang akan digunakan.
Database adalah struktur utama yang mengorganisasi data agar mudah diakses, dikelola, dan diolah.
Database harus dibuat terlebih dahulu sebelum membuat tabel apa pun.

<br>

## üìÇ 2. Melihat Daftar Database

Setelah MySQL dijalankan, pengguna dapat melihat daftar database yang tersedia.
Langkah ini penting untuk memastikan bahwa database yang dicari sudah ada atau belum dibuat.

<br>

## üéØ 3. Memilih Database yang Digunakan

Setelah database dibuat, langkah berikutnya adalah memilih atau mengaktifkan database tersebut.
Pemilihan database memastikan bahwa seluruh perintah berikutnya dijalankan pada database yang sama.
Hal ini mencegah kesalahan seperti menempatkan tabel pada database yang tidak diinginkan.

<br>

## ‚ùå 4. Menghapus Tabel atau Database

Pengguna dapat menghapus tabel atau seluruh database apabila sudah tidak diperlukan.
Proses ini bersifat permanen, sehingga harus dilakukan dengan hati-hati.
Data yang telah dihapus tidak dapat dipulihkan kembali.

<br>

## üö™ 5. Keluar dari MySQL CLI

Setelah selesai melakukan operasi, pengguna dapat keluar dari MySQL untuk menutup sesi kerja.
Langkah ini menandakan bahwa proses pengelolaan database sudah selesai dilakukan melalui CLI.

`EXIT;`

<br>

## Kesimpulan

Pengoperasian SQL melalui CLI memungkinkan kontrol penuh terhadap database tanpa menggunakan tampilan GUI.
Dengan CLI, pengguna dapat bekerja lebih cepat, akurat, dan efisien dalam mengelola database MySQL.
