## ðŸ§© Apa Itu ERD?

**Entity Relationship Diagram (ERD)** adalah diagram untuk memodelkan struktur awal sebuah database.

ERD membantu kita memahami:
- Data apa yang disimpan
- Objek apa saja yang terlibat
- Hubungan antar objek
- Atribut yang dimiliki tiap objek

## ðŸ”¹ 1. Entitas

**Entitas** adalah objek nyata atau konsep yang datanya ingin disimpan.

Contoh entitas:
- Mahasiswa
- Produk
- Pelanggan
- Transaksi

**Simbol:** kotak (rectangle)

Atribut adalah karakteristik atau informasi milik entitas.

**Simbol:** oval dengan garis bawah (underline)

**Simbol:** belah ketupat (diamond)
